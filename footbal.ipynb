```python
import pandas as pd 
```

```python
#reading 1 csv file from the website
df_premier23 = pd.read_csv('https://www.football-data.co.uk/mmz4281/2324/E0.csv')
```

```python
df_premier23
```

```python
#Rename Columns
df_premier23.rename(columns={'FTHG' : 'Home_goals',
                             'FTAG': 'Away_goals'}, inplace=True)
```

```python
# Show Data Frame
df_premier23
```

```python
# Filter rows where Liverpool is either the home or away team
liverpool_stats = df_premier23[(df_premier23['HomeTeam'] == 'Liverpool') | (df_premier23['AwayTeam'] == 'Liverpool')]

# Display the filtered data
print(liverpool_stats)
```

```python
# Calculates the total number of goals scored
# Total goals scored when Liverpool is the Home team
home_goals = liverpool_stats[liverpool_stats['HomeTeam'] == 'Liverpool']['Home_goals'].sum()

# Total goals scored when Liverpool is the Away team
away_goals = liverpool_stats[liverpool_stats['AwayTeam'] == 'Liverpool']['Away_goals'].sum()

# Total goals scored
total_goals_scored = home_goals + away_goals

print(f'Total Goals Scored: {total_goals_scored}')
```

```python
# Filter for Liverpool matches
home_goals = liverpool_stats[liverpool_stats['HomeTeam'] == 'Liverpool']['Home_goals'].sum()
away_goals = liverpool_stats[liverpool_stats['AwayTeam'] == 'Liverpool']['Away_goals'].sum()

print(f'Home Goals: {home_goals}, Away Goals: {away_goals}')
```

```python
# goals against
total_goals_against = liverpool_stats['Home_goals'][liverpool_stats['AwayTeam'] == 'Liverpool'].sum() + \
                      liverpool_stats['Away_goals'][liverpool_stats['HomeTeam'] == 'Liverpool'].sum()
print(f'Total Goals Against: {total_goals_against}')
```
